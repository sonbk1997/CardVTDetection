```python
def plot_bboxes(img_file: str, annot_file: str, class_dict: dict):
    image = cv2.imread(img_file)
    img_h, img_w, _ = image.shape
    print(img_h, img_w)

    with open(annot_file, 'r') as f:
        data = f.read().split('\n')
        data = [i.split(' ') for i in data[:]]
        data = [[float(j) for j in i] for i in data]

    # Calculating the bbox in Pascal VOC format
    for bbox in data:
        class_idx, x_center, y_center, w, h = bbox
        x_center = x_center*img_w
        y_center = y_center*img_h
        w = w*img_w
        h = h*img_h
        print(x_center, y_center, w, h)

        x_left_top = x_center - w/2
        y_left_top = y_center - h/2
        x_right_top = x_center + w/2
        y_right_top = y_center - h/2
        x_left_bottom = x_center - w/2
        y_left_bottom = y_center + h/2
        x_right_bottom = x_center + w/2
        y_right_bottom = y_center + h/2
        # print(x_left_top, y_left_top, x_right_top, y_right_top, x_left_bottom, y_left_bottom, x_right_bottom, y_right_bottom)
        pts = np.array([[x_left_top, y_left_top], [x_right_top, y_right_top], [x_right_bottom, y_right_bottom],
                [x_left_bottom, y_left_bottom]],
               np.int32)
        cv2.polylines(image, [pts], True, (255, 0, 0), 2)
    # Displaying the image

    cv2.imwrite('result.jpg', image)
    plt.imshow(image)
    plt.axis(False)
```
